# **Tuplas**

**Las tuplas son inmutables**

Una tupla es una secuencia de valores muy parecida a una lista. Los valores almacenados en una tupla pueden ser de cualquier tipo, y están indexados por números enteros. Las tuplas también son comparables y hashables (Un objeto es hashable si tiene un valor hash que nunca cambia durante su vida útil), por lo que se pueden ordenar y utilizarlas como valores clave en los diccionarios. Sintácticamente, una tupla es una lista de valores separada por comas.


```python
#Creamos la tupla
t = 'a', 'b', 'c', 'd', 'e'

#veamos que sea tipo tupla
print(type(t))

#imprimimos la tupla
print(t)
```

```python
#Creamos la tupla de otra forma, verificamos que sea tupla e imprimimos
t=('a', 'b', 'c', 'd', 'e')
print(type(t))
print(t)
%reset
```

```python
#Se busca obtener los valores desde la posicion 0 hasta la 3
t=('a', 'b', 'c', 'd', 'e')
t[0:3]
```

Aunque no es necesario, es común encerrar las tuplas entre paréntesis () para identificar las tuplas cuando se escribe código en Python:

```python
#Ejemplo de varios tipos de datos en una tupla
t1 = ('esto es una cadena', 123, 'c', 20.8, 'e',(32.2,32),[89,"Hola"])
print(t1)
t1[0]
```

# [] ---> Operador bracket
t1[0:3]

El anterior nos sirva para obtener un dato de dada posicion

```python
#En este caso vamos a obtener el valor en la posicion 0 del dato que esta en la posicion 5
print(t1[5][0])
```

```python
#En este caso vamos a obtener el valor en la posicion 3 del dato que esta en la posicion 1, del dato que esta en la posicion 6
print(t1[6][1][3])
```

Para crear una tupla con un solo elemento, hay que incluir la coma final:

```python
#Creando e imprimiendo la tupla de un solo elemento
t1 = ('a',)
type(t1)
print(t1)
```

Sin la coma Python trata ('a') como una expresión con una cadena entre paréntesis que evalúa a una cadena:

```python
#veamos que la trata como una cadena
t2 = ('a')
type(t2)
```

Otra forma de construir una tupla es la tupla de función incorporada. Sin ningún argumento, crea una tupla vacía:

```python
#Creando una tupla vacia
t = tuple()
print(type(t))
print(t)
```

Si el argumento es una secuencia (cadena, lista o tupla), el resultado de la llamada a la tupla es una tupla con los elementos de la secuencia:

```python
#Veamos como la tupla trata a las cadenas
t = tuple('Gestión de Datos')
print(t)
```

La mayoría de los operadores de listas también trabajan con tuplas. El operador bracket [] indexa un elemento:

```python
print(t[1:3])
```

Pero si intenta modificar uno de los elementos de la tupla, obtiene un error:

```python
#Codigo con error
#t[0] = 'A'
```

No se pueden modificar los elementos de una tupla, pero se puede reemplazar una tupla por otra:

```python
#reemplazando una tupla por otra
t = ('A',) + t[1:]
print(t)
```

# **Métodos**
count() Este método recibe un elemento como argumento, y cuenta la cantidad de veces que aparece en la tupla.


```python
#Contanto las t
t.count('t')
```

index()
Comparte el mismo método index() del tipo lista. Este método recibe un elemento como argumento, y devuelve el índice de su primera
aparición en la tupla.

```python
#Obteniendo el primer indice donde encuentre una e
t.index('e')
```
