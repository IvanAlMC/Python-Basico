# **Operadores Aritmeticos**

Tal como cualquier otro lenguaje de programacion, los operadores de suma, resta, multiplicación y división pueden ser usadas con números.
Con los números se pueden realizar los siguientes tipos de operaciones:
(+) Suma (-) Resta (*) Multiplicación
(/) División (//) División entera (**) Potencia

Ejemplo:


```python
a=input() #Recibimos un datos por teclado y lo guardamos en la variable a, teniendo en cuenta que este valor lo guarda como string por defecto
print(type(a)) # Imprime el tipo de dato de la variable con la función type(). Esto mostrará si a es una cadena (str), un número entero (int), u otro tipo de dato.
print(a+'3') #  Se concatena la cadena almacenada en a con la cadena '3'. Dado que a es una cadena y '3' también es una cadena, el operador + se utiliza para concatenarlas, es decir, unirlas en una sola cadena
print(int(a)+3) # Se intenta convertir la cadena a en un número entero utilizando int(a). Luego, se suma 3 al valor convertido.
print(a+str(3)) # Toma la cadena a y la concatena con la cadena resultante de convertir el número 3 en una cadena usando str(3).
```

# **Operadores de Identidad**

Se utilizan para comprobar si dos variables son, o no, el mismo objeto.

**is**

Devuelve True si las dos variables tienen almacenado el mismo valor.

**is not**

Devuelve True si las dos variables no tienen almacenado el mismo valor.

```python
x=y=1 # asignando el valor 1 a las variables x y y al mismo tiempo.
x is y # se utiliza el operador is para verificar si x y y apuntan al mismo objeto en la memoria.
```

```python
x=y=1 # asignando el valor 1 a las variables x y y al mismo tiempo.
x is not y # se utiliza el operador is para verificar si x y y no apuntan al mismo objeto en la memoria.
```

```python
w=3
z=3
w is z
```

```python
w=3
z=3
w is not z
```

Ejercicio de la formula cuadratica

```python
import math

# Coeficientes de la ecuación cuadrática
a = input()
b = input()
c = input()

# Calcular el discriminante
discriminante = int(b)**2 - 4*int(a)*int(c)

# Comprobar si el discriminante es positivo, negativo o cero
if discriminante > 0:
    # Dos soluciones reales distintas
    x1 = (-int(b) + math.sqrt(discriminante)) / (2*int(a))
    x2 = (-int(b) - math.sqrt(discriminante)) / (2*int(a))
    print("Las soluciones son x1 =", x1, "y x2 =", x2)
elif discriminante == 0:
    # Una solución real (raíz doble)
    x1 = -int(b) / (2*int(a))
    print("La solución doble es x =", x1)
else:
    # No hay soluciones reales
    print("No hay soluciones reales")
```
