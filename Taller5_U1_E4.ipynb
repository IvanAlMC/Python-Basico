# **Listas**

Se usan brackets para definir una lista y los elementos van separandos por comas.
La lista en Python es equivalente a un array, internamente cada posición puede ser de tipos de datos distintos.
Las listas en Python pueden son mutables, es decir, los elementos pueden modificarse. Una lista en Python es una estructura de datos formada por una secuencia ordenada de elementos.
Para acceder a los elementos de una lista se utilizan los indices, siendo 0 el índice del primer elemento.

```python
#Creando una lista de datos diferentes
lista = [1, 2.5, 'SQL/PLSQL', [5,6] ,4]
print(lista[0]) #imprimimos el datos de la posicion 0
print(lista[1]) #imprimimos el datos de la posicion 1
print(lista[2]) #imprimimos el datos de la posicion 2
print(lista[2][3]) #imprimimos el datos de la posicion 3 que esta en el dato de la posicion 2
print(lista[3]) #imprimimos el datos de la posicion 3
print(lista[3][0]) #imprimimos el datos de la posicion 0 que esta en el dato de la posicion 3
print(lista[3][1]) #imprimimos el datos de la posicion 1 que esta en el dato de la posicion 3
print(lista[1:3]) #imprimimos los datos desde la posicion 1 hasta la 2 (3-1)
print(lista[1:5]) #imprimimos el datos de la posicion 1 hasta la 4 (5-1)
```

Métodos
A continuación se presentan algunos métodos integrados en Python:

append()
Este método agrega un elemento al final de una lista.

```python
%reset
#Se crea una nueva lista
distribuciones_linux=['Ubuntu', 'CentOS8.0','Debian','Debian']
print(distribuciones_linux)
#Agregamos el nuevo elemento
distribuciones_linux.append('Redhat')
print(distribuciones_linux)
# Calcular la cantidad de elementos de la lista
len(distribuciones_linux)
```

count()
Este método recibe un elemento como argumento, y cuenta la cantidad de veces que aparece en la lista.


```python
#Se cuenta cuantas veces esta 'Debian'
distribuciones_linux.count('Debian')
```

extend()
Este método extiende una lista agregando un iterable al final. Es equivalente al método append()

```python
distribuciones_linux.extend(['Oracle Linux8.0'])
distribuciones_linux
```

```python
# Observar la diferencia de agregar un elemento a la lista
distribuciones_linux.extend('Oracle Linux8.0')
print(distribuciones_linux)
```

```python
distribuciones_linux=['Ubuntu', 'CentOS8.0','Debian','Debian']
```

index()
Este método recibe un elemento como argumento, y devuelve el índice de su primera aparición en la lista.

```python
#Obtenemos el primer indice que tenda el dato 'Ubutu'
distribuciones_linux.index('Ubuntu')
```

insert()
Este método inserta el elemento x en la lista, en el índice i.

```python
distribuciones_linux.insert(2,'Linux Mint')
print(distribuciones_linux)
```

pop()
Este método devuelve el último elemento de la lista, y lo borra de la misma.

```python
#Se hace un pop
print(distribuciones_linux.pop())
print(distribuciones_linux)
```

sort()
Este método ordena los elementos de una lista.

```python
#Ordenando listas, en este caso alfabeticamente
distribuciones_linux.sort()
print(distribuciones_linux)
```

El método sort() admite la opción reverse, por defecto, con valor False. De tener valor True, el ordenamiento se hace en sentido inverso.


```python
#Ordenamos inversamente
distribuciones_linux.sort(reverse=True)
print(distribuciones_linux)
```

# **Usando listas como pilas**

Los métodos de lista hacen que sea muy fácil usar una lista como pila, donde la última El elemento agregado es el primer elemento recuperado ( “ último en entrar, primero en salir ” ). Para agregar un elemento en la parte superior de la pila, usar append(). Para recuperar un artículo del parte superior de la pila, usar pop() sin un índice explícito. Por ejemplo:

```python
#añadiendo elementos en forma de pila
stack = [3, 4, 5]
stack.append(6)
stack.append(7)
stack
```

```python
#obteniendo y eliminando elementos en forma de pila
stack.pop()
```

# **Usando listas como colas**

También es posible usar una lista como cola, donde se agrega el primer elemento el primer elemento recuperado ( “ primero en entrar, primero en salir ” ); sin embargo, las listas no son eficiente para este propósito. Mientras que los apéndices y los pops del final de la lista son rápido, hacer insertos o pops desde el comienzo de una lista es lento ( porque todo de los otros elementos deben ser desplazados por uno ).

Para implementar una cola, use collections.deque que fue diseñado para tener apéndices rápidos y pops de ambos extremos. Por ejemplo:

```python
#Creando una cola
from collections import deque
queue = deque(["Eric", "John", "Michael"])
#Añadiendo elementos al estilo de cola
queue.append("Terry")
queue.append("Graham")
queue
```

```python
#Obteniendo y eliminando elementos al estilo de cola
queue.popleft()
```
