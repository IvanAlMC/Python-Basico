# **CONJUNTOS o SET**

Colección de objetos no repetidos e inmutables. Estos conjuntos se utilizan en operaciones de lógica matemática. Al igual que otras colecciones.
Al igual que otras colecciones, sus miembros pueden ser de diversos tipos

```python
#Creando una coleccion, imprimiendola y verificando su tipo
s = {True, 3.14, None, False, "Hola mundo", (1, 2)}
print(s)
print(type(s))
```

No obstante, un conjunto no puede incluir objetos mutables como listas, diccionarios, e incluso otros conjuntos.


```python
# s = {1,{3,4}} codigo da error
# s = {[1, 2]} codigo da error
s = set()
print(s)
```

De la misma forma podemos obtener un conjunto a partir de cualquier objeto iterable

```python
s1 = set([1, 2, 2, 3, 4])
s2 = set(range(10))
s3 = set([4,6,4,2,5,1,3,2])
print(s1)
print(s2)
print(s3)
```

Crea un conjunto con una serie de elementos entre llaves Los elementos repetidos se eliminan

```python
set([1, 2, 2, 3, 4])
c = {1, 3, 2, 9, 3, 1}
print(c)
```

Crea un conjunto a partir de un string. Los caracteres repetidos se eliminan

```python
a = set("Hola Pythonista")
print(a)
```

```python
a="Hola Pythonista".upper()
print(a)
```

```python
a=set(a)
print(a)
```

```python
a = set("Hola Pythonista".upper())
print(a)
```

Ejercicio: consulte la diferencia entre set y frozenset

En Python, tanto set como frozenset son tipos de datos utilizados para almacenar colecciones de elementos sin duplicados, pero tienen una diferencia clave: la mutabilidad.

set:
set es una colección mutable, lo que significa que puedes agregar, eliminar y modificar elementos después de crear el conjunto.
Los conjuntos se definen utilizando llaves {} o utilizando el constructor set().
Los conjuntos no admiten elementos duplicados, por lo que automáticamente eliminan duplicados cuando los agregas.
Ejemplo de uso de set:

```python
mi_set = {1, 2, 3}
mi_set.add(4)  # Agregar un elemento
mi_set.remove(2)  # Eliminar un elemento
print(mi_set)  # Salida: {1, 3, 4}
```

frozenset:
frozenset es una colección inmutable, lo que significa que una vez que se crea un frozenset, no se pueden realizar cambios en su contenido.
Los frozensets se definen utilizando el constructor frozenset().

Ejemplo de uso de frozenset:

```python
mi_frozenset = frozenset([1, 2, 3])
# Intentar modificar el frozenset resultaría en un error:
# mi_frozenset.add(4)  # Esto genera un error
```

La principal ventaja de frozenset es su inmutabilidad, lo que significa que puedes usarlo como clave en un diccionario (ya que las claves deben ser objetos inmutables en un diccionario), mientras que no puedes hacerlo con un set. Por ejemplo:

```python
diccionario = {frozenset([1, 2, 3]): "valor"}
print(diccionario)  # Salida: {frozenset({1, 2, 3}): 'valor'}
```

En resumen, la diferencia clave entre set y frozenset radica en su mutabilidad. set es mutable y permite cambios en su contenido, mientras que frozenset es inmutable y no permite cambios después de su creación.

# **Cómo acceder a los elementos de un conjunto en Python?**

Dado que los conjuntos son colecciones desordenadas, en ellos no se puede referenciar a la posición de cada elemento, como ocurre en los tipos list o tuple. Es por ello que no se puede acceder a los elementos a través de un índice.

Sin embargo, sí se puede acceder y/o recorrer todos los elementos de un conjunto usando un bucle for:

```python
mi_conjunto = {1, 3, 2, 9, 3, 1}
print(mi_conjunto)
for numero in mi_conjunto:
  print(numero)

#print(mi_conjunto[0]) esta linea produce un error
```

# **Métodos**

Los objetos de tipo conjunto mutable y conjunto inmutable integra una serie de métodos integrados a continuación:

add()

Este método agrega un elemento a un conjunto mutable. Esto no tiene efecto si el elemento ya esta presente.

```python
#Añadiendo un dado a un conjuto mutable
set_mutable1 = set([4, 3, 11, 7, 5, 2, 1, 4])
print(set_mutable1)

set_mutable1.add(22.000001)
print(set_mutable1)
```

clear()
Este método remueve todos los elementos desde este conjunto mutable.

```python
set_mutable1.clear()
print(set_mutable1)
```

copy()
Este método devuelve una copia superficial del tipo conjunto mutable o conjunto inmutable:

```python
#Copiando un set en otro, y verificando si son iguales
set_mutable = set([4.0, 'Carro', True])
otro_set_mutable = set_mutable.copy()
set_mutable == otro_set_mutable
```

difference()
Este método devuelve la diferencia entre dos conjuntos, es decir, los elementos que están en el primero conjunto y no en el segundo. En operadores SQL es equivalente a minus.

```python
set_mutable1 = set([4, 3, 11, 7, 5, 2, 1, 4])
set_mutable2 = set([11, 5, 9, 2, 4, 8,2])
print(set_mutable1)
print(set_mutable2)
print(set_mutable1.difference(set_mutable2)) #Obteniendo los elementos del set_mutable 1 que no estan en set_mutable 2
print(set_mutable2.difference(set_mutable1)) #Obteniendo los elementos del set_mutable 2 que no estan en set_mutable 1
```

La diferencia también se puede generar con el operador menos (-), creando un nuevo conjunto.

```python
a = {1, 2, 3, 4}
b = {2, 3}
c = a - b
print(c)
```

Dos conjuntos son iguales si y solo si contienen los mismos elementos (a esto se lo conoce como principio de extensionalidad):

```python
{1, 2, 3} == {3, 2, 1}
```

difference_update()
Es el mismo operador difference o menos(-), pero actualizando uno de los conjunstos.


```python
proyecto1 = {'python', 'Zope2', 'ZODB3', 'pytz'}
print(proyecto1)
proyecto2 = {'python', 'Plone', 'diazo'}
print(proyecto2)
#Se actualiza el proyecto 2 con con los valores que no estan en proyecto 1
proyecto2.difference_update(proyecto1)
print(proyecto2)
```

discard()
Este método remueve un elemento desde un conjunto si esta presente.

```python
paquetes = {'python', 'zope', 'plone', 'django'}
print(paquetes)
paquetes.discard('django')
print(paquetes)
```

intersection()
Este método devuelve la intersección entre los conjuntos.

```python
set_mutable1 = set([4, 3, 11, 7, 5, 2, 1, 4])
set_mutable2 = set([11, 5, 9, 2, 4, 8])
print(set_mutable1)
print(set_mutable2)
print(set_mutable1.intersection(set_mutable2))
print(set_mutable2.intersection(set_mutable1))
```

También se puede expresar con el operador &, y retorna un nuevo conjunto con los elementos que se encuentran en ambos.


```python
set_intersection = set_mutable1 & set_mutable2
print(set_intersection)
```

# ***CONSULTA***

Utilice los siguientes operadores formulando un ejemplo: intersection_update() isdisjoint() issubset() issuperset() pop() remove()
symmetric_difference() symmetric_difference_update() union() update()

```python
# Definir dos conjuntos
conjunto1 = {1, 2, 3, 4, 5}
conjunto2 = {4, 5, 6, 7, 8}

print(conjunto1)
print(conjunto2)

# intersection_update() - Actualiza conjunto1 con la intersección de conjunto1 y conjunto2
conjunto1.intersection_update(conjunto2)
print(conjunto1)  # Resultado: {4, 5}
```

```python
# isdisjoint() - Comprueba si dos conjuntos son disjuntos (no tienen elementos en común)
son_disjuntos = conjunto1.isdisjoint(conjunto2)
print(son_disjuntos)  # Resultado: False (porque tienen elementos en común)
```

```python
# issubset() - Comprueba si conjunto1 es un subconjunto de conjunto2
es_subconjunto = conjunto1.issubset(conjunto2)
print(es_subconjunto)  # Resultado: True (porque conjunto1 es un subconjunto de conjunto2)
```

```python
# issuperset() - Comprueba si conjunto1 es un superconjunto de conjunto2
es_superconjunto = conjunto1.issuperset(conjunto2)
print(es_superconjunto)  # Resultado: False (porque conjunto1 no es un superconjunto de conjunto2)
```

```python
# pop() - Elimina y devuelve un elemento al azar del conjunto
elemento_eliminado = conjunto1.pop()
print(elemento_eliminado)  # Resultado: Un elemento aleatorio que se elimina de conjunto1
```

```python
# remove() - Elimina un elemento específico del conjunto
conjunto1.add(4)
conjunto1.remove(4)
print(conjunto1)  # Resultado: {5}
```

```python
# symmetric_difference() - Devuelve un nuevo conjunto con elementos que están en uno de los conjuntos, pero no en ambos
diferencia_simetrica = conjunto1.symmetric_difference(conjunto2)
print(diferencia_simetrica)  # Resultado: {4, 6, 7, 8}
```

```python
# symmetric_difference_update() - Actualiza conjunto1 con la diferencia simétrica entre conjunto1 y conjunto2
conjunto1.symmetric_difference_update(conjunto2)
print(conjunto1)  # Resultado: {5, 6, 7, 8}
```

```python
# union() - Devuelve un nuevo conjunto con todos los elementos de ambos conjuntos
union_conjuntos = conjunto1.union(conjunto2)
print(union_conjuntos)  # Resultado: {4, 5, 6, 7, 8}
```

```python
# update() - Agrega elementos de un conjunto a otro
conjunto1.update(conjunto2)
print(conjunto1)  # Resultado: {4, 5, 6, 7, 8}
```
