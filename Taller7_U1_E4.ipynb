# **Diccionarios**

El diccionario, define una relación uno a uno entre una clave y su valor.
Los diccionarios en Python, al igual que las listas y las tuplas, permiten almacenar diferentes tipos de datos: Strings, enteros, flotantes, booleanos, tuplas, listas e inclusive otros diccionarios.
Los diccionarios son mutables, es decir, es posible modificar su longitud, podemos agregar o quitar elementos de él; de igual forma todos los valores almacenados en el diccionario pueden ser modificados.


Definición y almacenamiento de datos
Para definir un diccionario se utilizan corchetess o la función dict(). De ambas formas es correcto.

```python
diccionario={}
print(diccionario)
print(type(diccionario))
diccionario = dict()
diccionario
```

Para almacenar algún valor se utiliza la siguiente estructura: los corchetes, dos puntos y el valor el cual se quiere asociar.
Por ejemplo, el string total, dos puntos, 55, con esto se le indica a Python que la llave "total" almacena el valor 55.

```python
diccionario = {"total": 55}
print(diccionario)
```

Para definir un diccionario con varios valores utilice la coma para separarlos.

```python
diccionario = {"total": 55, "descuento": True, "subtotal": 15}
print(diccionario)
```

Otro ejemplo:

```python
diccionario = {"total": 55, 10: "Curso de Python", 2.0: True}
print(diccionario)
```

# **Métodos**
Los objetos de tipo diccionario integran una variedad de métodos, como por ejemplo:

clear() Este método borra todos los elementos del diccionario.

```python
versiones = dict(python=2.7, zope=2.13, plone=5.1)
print(versiones)
```

```python
versiones.clear()
print(versiones)
```

copy() Este método devuelve una copia del diccionario:

```python
versiones = dict(python=2.7, zope=2.13, plone=5.1)
otro_versiones = versiones.copy()
print(versiones == otro_versiones)
print(type(otro_versiones))
```

fromkeys() Este método crea un nuevo diccionario con claves a partir de un tipo de dato secuencia. El valor de value por defecto es el tipo None.

```python
secuencia = ('python', 'zope', 'plone')
versiones = dict.fromkeys(secuencia)
print("Nuevo Diccionario : %s" % str(versiones))
print (versiones)
```

En el ejemplo anterior inicializa los valores de cada clave a None, también se puede inicializar con un valor común por defecto para cada clave:

```python
versiones = dict.fromkeys(secuencia, 0.1)
print("Nuevo Diccionario : %s" % str(versiones))
```

items()
Este método devuelve una lista de pares de diccionarios (clave, valor), representadas en tuplas.

```python
versiones = dict(python=2.7, zope=2.13, plone=5.1)
print(versiones)
print(versiones.items())
```

len() Esta función es la misma función integrada len() en el interprete Python pero aplicada a la estructura de datos "diccionario".

```python
versiones = dict(python=2.7, zope=2.13, plone=5.1)
len(versiones)
```

Eliminar un elemento de un diccionario.


```python
#Con el metodo pop
versiones.pop('zope')
versiones.items()
```

```python
#Con del
del versiones['plone']
versiones.items()
```

UPDATE() pega y/o Actualiza diccionarios

```python
diccionario1 = {"color": "verde", "precio": 45}
diccionario2 = {"talle": "M", "marca": "Lacoste"}
diccionario3 = {"talle": "L","almacen" : "Olimpica"}
diccionario1.update(diccionario2)
print(diccionario1)
diccionario1.update(diccionario3)
print(diccionario1)
```

SETFEFAULT() para definir Llaves y Valores por defecto

```python
dic1 = {"color": "rosa", "marca": "Zara"}
clave = dic1.setdefault("talle", "U")
print(clave)
print(dic1)
clave2 = dic1.setdefault("color", "verde")
print(clave2)
print(dic1)
```

ENUMERATE Enumerate es una función incorporada de Python. Permite recorrer una estructura de datos y asignar un contador automático. La estructura es: for counter, value in enumerate(some_list): print(counter, value)


```python
#enumerando una iteracion
my_list = ['apple', 'banana', 'grapes', 'pear']
for c, value in enumerate(my_list, 1):
  print(c, value)
```

```python
my_list = ['apple', 'banana', 'grapes', 'pear']
counter_list = list(enumerate(my_list, 1))
print(counter_list)
```

```python
some_list = ['foo', 'bar', 'baz']
mapping = dict((k, v) for v, k in enumerate(some_list))
print(mapping)
```

Construcción de Diccionarios a partir de dos listas utilizadon ZIP Function.

¿Qué es el método zip en Python?
La función zip() devuelve un objeto zip, que es un iterador de colecciones, donde el primer elemento de cada iterador pasado se empareja y luego el segundo elemento de cada iterador pasado se empareja.

```python
key_list=['id','name','salary']
value_list=[10,'Hugo',31.34]
empleado = {}
for key, value in zip(key_list, value_list):
  empleado[key] = value
print(empleado)
value = empleado.get('name')
print(value)
```

Actividad Crear un diccionario a partir de una lista, clasificando las palabras que inician con la misma letra. Como se muestra en el siguiente ejemplo:
Lista--> words = ['apple', 'bat', 'bar', 'atom', 'book','cat']
Diccionario de salida--->{'a': ['apple', 'atom'], 'b': ['bat', 'bar', 'book'], 'c': ['cat']}

```python
words = ['apple', 'bat', 'bar', 'atom', 'book', 'cat']
# Inicializamos un diccionario vacío
word_dict = {}

# Iteramos sobre la lista de palabras
for word in words:
    # Obtenemos la primera letra de la palabra (clave)
    first_letter = word[0]
    # Verificamos si la clave ya existe en el diccionario
    if first_letter in word_dict:
        # Si existe, agregamos la palabra a la lista de esa clave
        word_dict[first_letter].append(word)
    else:
        # Si no existe, creamos una nueva lista con la palabra como primer elemento
        word_dict[first_letter] = [word]

# El diccionario final contendrá las palabras agrupadas por la letra inicial
print(word_dict)
```
