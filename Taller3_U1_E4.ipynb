# **Seleccionar un fragmento de un String**

Un segmento de un string se llama un trozo. Seleccionar una rebanada es similar a seleccionar más de un caracter:


```python
s='Gestión de Datos'
print(s[0:5])
```

```python
print(s[8:11])
```

El operador devuelve la parte de la cadena del "n-ésimo" carácter "m-ésimo" al carácter, incluyendo el primero pero excluyendo el último [n,m). Si se omite el primer índice (antes de los dos puntos), el trozo comienza al principio de la cadena. Si se omite el segundo índice, la rebanada va al final de la cadena:

```python
print(s[:3])
```

```python
print(s[3:])
```

Si el primer índice es mayor o igual que el segundo el resultado es una cadena vacía, representada por dos comillas:

```python
print(s[3:3])
```

El método format() de string Se pueden definir variables y poner corchetes {} en el string donde van a ir los números o caracteres. Entre paréntesis pasamos los valores. Unos ejemplos sencillos.


```python
# saca "El valor es 69
print ("El valor es {}".format(69))
```

```python
# saca "El valor es 12.3456
print ("El valor es {}".format(12.3456))
```

```python
# Tres conjuntos {}, el primero para el primer parámetro de format(),
# el segundo para el segundo
# y así sucesivamente.
# saca "Los valores son 1, 2 y 3"
print ("Los valores son {}, {} y {}".format(1,2,3))
```

```python
# Entre las llaves podemos poner la posición del parámetro.
# {2} es el tercer parámetro de format()
# {0} es el primer parámetro de format.
# saca "Los valores son 3, 2 y 1"
print ("Los valores son {2}, {1} y {0}".format(1,2,3))
```

El método de división de una cadena según un caracter separador más empleado es split(), cuyo separador por defecto son espacios en blanco y saltos de línea.

```python
"Hola mundo!\nHello world!".split()
```

El separador puede indicarse como argumento.

```python
"Hola mundo!\nHello world!".split(" ")
```

O bien, para separar únicamente según saltos de línea:

```python
"Hola mundo!\nHello world!".splitlines()
```

Un segundo argumento en split() indica cuál es el máximo de divisiones que puede tener lugar (-1 por defecto para representar una cantidad ilimitada).

Un segundo método de separación es partition(), que retorna una tupla de tres elementos: el bloque de caracteres anterior a la primera ocurrencia del separador, el separador mismo, y el bloque posterior.

```python
s = "Hola mundo. Hello world!"
s.partition(" ")
```

rpartition() opera de forma similar, pero realizando la búsqueda de derecha a izquierda.

```python
s.rpartition(" ")
```

Por último, el método join() ─sumamente útil─, que debe ser llamado desde una cadena que actúa como separador para unir dentro de una misma cadena resultante los elementos de una lista.

Como puede observarse, split() y join() son exactamente contrarios.

Almacenar una cadena de caracteres en una variable y evaluar cuantas vocales tiene la frase.
Ejemplo: "Gestión de Datos"
Salida: a --> 1 e ---> 1 i --- > 1 o ---> 2 u ---> 0 Otras letras --->

```python
def contar_vocales(frase):
    # Convierte la frase a minúsculas para considerar tanto mayúsculas como minúsculas
    frase = frase.lower()

    # Inicializa contadores para cada vocal
    contador_a = contador_e = contador_i = contador_o = contador_u = 0

    # Recorre cada carácter en la frase
    for letra in frase:
        if letra == 'a':
            contador_a += 1
        elif letra == 'e':
            contador_e += 1
        elif letra == 'i':
            contador_i += 1
        elif letra == 'o':
            contador_o += 1
        elif letra == 'u':
            contador_u += 1

    # Calcula el número de "otras letras" restando las vocales del total de letras
    otras_letras = len(frase) - (contador_a + contador_e + contador_i + contador_o + contador_u)

    # Devuelve un diccionario con los resultados
    resultados = {
        'a': contador_a,
        'e': contador_e,
        'i': contador_i,
        'o': contador_o,
        'u': contador_u,
        'Otras letras': otras_letras
    }

    return resultados

# Ejemplo de uso
frase = "Gestión de Datos"
resultado = contar_vocales(frase)

# Imprime el resultado
for vocal, cantidad in resultado.items():
    print(f'{vocal} --> {cantidad}')
```
